```python
Loan_Main = data[~data['loan_status'].isin(['Current'])]
Loan_Main["loan_status"] = np.where(Loan_Main["loan_status"].str.contains("Fully Paid"), 0, 1)

# New Dataset shape
print(Loan_Main.shape)

# Changing employee length to numerical
Loan_Main['emp_length'].fillna("< 1 year", inplace=True)
dict_emp_length = {'10+ years': 10, '6 years': 6, '4 years': 4, '< 1 year': 0.5, '2 years': 2, '9 years': 9, 
                   '5 years': 5, '3 years': 3, '7 years': 7, '1 year': 1, '8 years': 8}
Loan_Main['emp_length'] = Loan_Main['emp_length'].replace(dict_emp_length)

# Separating Loan_Main dataset into two based on Application_Type (Individual / Joint Type)
Individual_Loan = Loan_Main[Loan_Main['application_type'] == "Individual"]
Joint_Loan = Loan_Main[Loan_Main['application_type'] == "Joint App"]
```

```python
# List of columns to drop
irrelevant_columns = ['id', 'emp_title', 'desc', 'title', 'zip_code', 'url']

# Drop these columns if they exist in the DataFrame
df = df.drop(columns=irrelevant_columns, errors='ignore')

# Replace specific string values with NaN or appropriate numerical value
df = df.replace({'Total amount funded in policy code 1: 6417608175': np.nan})

# Label Encoding for binary and ordinal categorical variables
label_encoder = LabelEncoder()
binary_columns = ['term', 'pymnt_plan', 'initial_list_status', 'hardship_flag', 'debt_settlement_flag']
ordinal_columns = ['grade', 'sub_grade', 'verification_status', 'verification_status_joint']

for col in binary_columns:
    if col in df.columns:
        df[col] = label_encoder.fit_transform(df[col])

for col in ordinal_columns:
    if col in df.columns:
        df[col] = label_encoder.fit_transform(df[col])

# One-Hot Encoding for nominal categorical variables
nominal_columns = [col for col in non_numeric_columns if col not in binary_columns + ordinal_columns + irrelevant_columns]
df = pd.get_dummies(df, columns=nominal_columns, drop_first=True)

# Display the first few rows of the processed DataFrame
print(df.head())

# Save the processed DataFrame for further use
df.to_csv('processed_dataset.csv', index=False)
```
